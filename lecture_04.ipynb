## Решение нелинейных  алгебраических уравнений и систем. Поиск экстремумов

### Постановка задачи

Дано нелинейное алгебраическое уравнение
$$
F(x)=0
$$

Нелинейность уравнения означает, что график функции не прямая линия. 

Решить уравнение - это найти 
$$
x^*\subset R:F(x^*)=0
$$

Значение x* называют корнем уравнения. Нелинейное уравнение может иметь несколько корней. 

Геометрическая интерпретация корней уравнения представлена

на рисунке

![формула трапеций](images/005.png)

Корнями уравнения являются точки $x_1^*\,$, $x_2^*\,$, $x_3^*\,$,

в которых функция $F(x)$ пересекает ось $x$.

Необходимое условие существования корня уравнения и достаточное условие единственности 

следуют из известной теоремы Больцано–Коши. Пусть $F(x)$ непрерывна и $F(a)F(b)<0$ (т.е. на 

концах интервала функция имеет разные знаки). Тогда внутри отрезка $[a,b]$ существует корень 

уравнения $F(x)=0$. Корень будет единственным, если $F'(x)$ не меняет знак на отрезке 

$[a,b]$, т.е. $F(x)-$ монотонная функция.

Методы решения уравнения (1.1) можно разделить на точные  (аналитические) и приближенные 

(итерационные). В точных методах корень находится за конечное число действий и 

представляется некоторой алгебраической формулой. 

В приближенных методах процесс нахождения решения бесконечен. 

Приближенным решением рассматриваемого уравнения называется бесконечная последовательность 

${x_n}$ такая, что $lim_{n\to\infty}{x_n}=x^*$. По определению предела,  для любого (сколь 

угодно малого) $\varepsilon$, найдется такое $N$, что при любом $n>N$,

$|x_n-x^*|<\varepsilon$. Члены этой последовательности $x_n$ называются последовательными 

приближениями к решению, или итерациями.

Наперед заданное число $\varepsilon$ - точность метода, а $N$ - это количество итераций, 

которое необходимо выполнить,  чтобы получить решение с заданной точностью $\varepsilon$. 


### Скалярные уравнения (одномерный случай)


###### Первая задача: (метод грубой силы)

1) Требуется найти все корни скалярного алгебраического уравнения
$$
    f(x) = 0
$$
на отрезке $[a,b]$.


Предположим, что мы имеем множество точек вдоль кривой функции f (x) на отрезке $[a,b]$:

![формула трапеций](images/006.png)

Мы хотим решить $f (x) = 0$, т. е. найти точки $x$, где $f(x)$ пересекает

 ось $x$. 

Алгоритм грубой силы должен проходить через все точки кривой

и проверьте, находится ли одна точка ниже оси x, и если следующая точка выше

ось x или наоборот. Если это будет так, мы получим 

что значение $f(x)$ должно быть равным нулю между этими двумя точками x.

Численный алгоритм.

Мы имеем множество $n+1$ точек $(x_i,y_i)$

$$
y_i=f(x_i),\,\,i=0,.....,n
$$
и $x$ изменяется
$$
0<x_1<...<x_n
$$

Проверяем условие, если $y_i<0$ и $y_{i+1}>0$ (или наоборот). Более компактное выражение для 

этой проверки выглядит так 
$$
y_i\cdot y_{i+1}<0
$$
Если это так, то корень $f(x)\varepsilon=0$ находится в $[x_i,x_{i+1}]$, и мы по двум этим 

точкам проводим наклонную прямую и мы имеем приближение


$$
f(x)\approx \frac{f(x_{i+1})-f(x_i)}{x_{i+1}-x_i}\cdot (x-x_i)+f(x_i)=\frac{y_{i+1}-y_i}{x_{i+1}-x_i}\cdot (x-x_i)+y_i
$$


и для $x$ получаем формулу

$$
x=x_i-\frac{x_{i+1}-x_i}{y_{i+1}-y_i}\cdot y_i
$$

Таким образом рассматриваем все узлы нашей равномерной сетки и находим все корни нелинейного 

уравнения.

Абсолютная погрешность вычисляется по формуле

$$
\Delta=\frac{b-a}{n}
$$

Чем меньше сетка разбиения, тем точней метод.


##### Вторая задача


2) Требуется найти экстремумы скалярной алгебраической функции $f(x)$ на отрезке $[a,b]$.

Мы понимаем, что $x_i$ соответствует максимум, если выполняется условие 

$$
y_{i_1}<y_i>y_{i+1}
$$

Аналогично, если $x_i$ соответствует минимум, по будет выполнятся условие

$$
y_{i-1}>y_i<y_{i+1}
$$

Так же мы должны понимать, мы так можем найти все локальные максимумы и минимумы для всех 

"внутренних" точек $i=1,2,....n-1$ рассматриваемого отрезка. Кроме того, нам нужно добавить 

концы отрезка $i=0$ или $i=n$, если соответствующий $y_i$ является глобальным максимумом или

минимумом.\approx 




#### Метод Ньютона нахождения корней нелинейных уравнений


Метод Ньютона, также известный как метод Ньютона-Рафсона, очень известный и широко 

используемый метод решения нелинейных алгебраических уравнений. По сравнению с другими 

методами, как правило, самый быстрый. Однако этот метод не гарантирует нахождения корня уравнения.

Основная идея численных методов нелинейного уравнения состоит в том, что мы находим линейное

уравнение (поскольку мы знаем, как решать линейное уравнение) и надеемся, что решение этого

линейного уравнения приблизит нас к решению нелинейного уравнения. 

Эта идея становится более ясной, когда мы рассмотрим метод Ньютона и секущий метод.


![формула трапеций](images/008.png)

Рассмотрим нелинейное алгебраические уравнение $f(x)=0$.

На рисунке показана функция $f(x)$.

Численные методы для алгебраических уравнений требуют от нас угадать начало (с какого узла 

начинать приближение). Здесь это предположение называется $x_0$.

Основная идея метода Ньютона состоит в том, чтобы аппроксимировать исходную функцию $f(x)$

прямой линией, т.е. заменить линейной функцией, та как легко решить линейное уравнение.

Метод Ньютона, в качестве линейной функции выбирает касательную к $f(x)$. 

Если посмотреть на предложенный график, то начало начинается с точки $x_0$. И в этой точке

строится первая касательная к графику функции $f(x)$. Самая правая касательная на графике.

Эта касательная пересекает ось $0x$ в точке, которую мы назавем $x_1$.

Это, мы надеемся, лучшее приближение функции $f(x)=0$, чем в точке $x_0$.

Следующий шаг: опять строим касательную в уже в точке $x_1$ и опять получим точку $x_2$, 

которая даст еще лучшей результат, чем в точке $x_1$  и так идем до корня уравнения.

Уравнение касательной имеет вид:

$$
F(x)=f(x_0)+f'(x_0)(x-x_0)
$$
Решаем уравнение 
$$
f(x)=0
$$
Приближаем методом Ньютона, следовательно

$$
f(x)=f(x_0)+f'(x_0)(x-x_0)=0
$$

Находим $x_1$

$$
x_1=x_0-\frac{f(x_0)}{f'(x_0)}
$$

Вычисления повторяются до тех пор, пока $f(x)$ не будет достаточно близко к нулю. 

Точнее, мы проверяем, если $|f(x_n)|<\varepsilon$. 

$\varepsilon$ - точность метода.


#### Метод секущих нахождения корней нелинейных уравнений

Когда мы не можем вычислить производную функции, например функция задана таким видом 

$$
f(x)=\int_0^x{e^{-t^2}dt}
$$

метод Ньютона становится проблемой. Тогда мы можем использовать другой метод - метод секущих

для нахождения корней нелинейных уравнений. 

Вместо использования касательных линий к графику, мы можем использовать секущие. Этот подход

называется секущим методом, и идея проиллюстрирована графически на следующем рисунке.

![формула трапеций](images/009.png)


Идея секулярного метода состоит в том, чтобы думать, как в методе Ньютона, но вместо того, 

чтобы использовать $f'(x_n)$, мы аппроксимируем эту производную секущей.

$$
f'(x)\approx \frac{\Delta y}{\Delta x}=\frac{y_{n}-y_{n-1}}{x_{n}-x_{n-1}}=\frac{f(x_n)-f(x_{n-1})}{x_{n}-x_{n-1}}
$$

Используя формулу для нахождения $x_i$? где $i=0,...n$ из метода Ньютона, мы получим:

$$
x_{n+1}=x_n-\frac{f(x_n)}{\frac{f(x_n)-f(x_{n-1})}{x_{n}-x_{n-1}}}=x_n-f(x_n)\cdot\frac{x_n-x_{n-1}}{f(x_n)-f(x_{n-1})}
$$

Посмотрим на график, мы видим, как две выбранные начальные точки ($x_0=1000,\,\,\,x_1=700$ и 

соответствующие значения функции) используются для вычисления $x_2$. Как только мы получаем

$x_2$, мы так же используем $x_1,\,\,x_2$ для нахождения $x_3$. Как и в случае с методом 

Ньютона, процедура повторяется до тех пор, пока $f(x_n)$ не будет достаточно близко к нулю. 

Точнее, мы проверяем, если $|f(x_n)|<\varepsilon$. 

$\varepsilon$ - точность метода.


#### Метод бисекции

Ни метод Ньютона, ни секущий не могут гарантировать, существующее решение будет найдено.

Способ бисекции или метод деления отрезка пополам это решение находит. Однако, если у 

уравнения есть несколько решений -то этот метод находит только один из них, точно так же, 

как метод Ньютона и метод секущих. Метод бисекции медленнее, чем два других метода. Поэтому 

надежность метода приходит со скоростью.

Идея метода заключается во первых, что рассматривается непрерывная функция на интервале и 

значения функции на концах интервала имеет противоположные знаки. Во вторых - ключевая идея 

-это разделение интервала на два равных интервала, один слева и один справа от середины 

интервала. Оценив знак в серединной точке, мы узнаем сразу решение должно существовать слева 

или справа от середины интервала или мы уже нашли корень (если значение функции обратилось в 

ноль).

Рассмотрим метод на примере уравнения $x^2-9=0,\,\,x\in [a,b]=[0,100]$.

Первое условие метода - непрерывность функции на отрезке выполнено.

И функция имеет противоположные знаки на концах отрезка.

$$
f(x)=x^2-9
$$
$$
f(0)=-9\,\,f(100)=9991
$$

Значит есть хоть одно решение уравнения.

Поделили пополам получили $x=500$.

Оценили знак функции $f(500)\gt 0$. Следовательно корень уравнения должен находится между 

$a$ - началом отрезка и серединой отрезка. Функция пересекает ось $ox$ не реже одного раза.

В любом случае мы продолжим только половину интервала. Исключение, если $f(x)=0$ в 

серединной точке. Следующий шаг опять деление отрезка и проверка значения функции на концах. 

И так до тех пор пока не найдется корень уравнения. "Решением" в этом случае будет тогда, 

когда выполнено условие $|f(x)|<\varepsilon$. 







### Решение системы нелинейных уравнений методом Ньютона

В этой лекции мы рассмотрели одно линейное алгебраическое уравнение. Однако системы таких

уравнений возникают в ряде приложений, прежде всего нелинейных обыкновенных и 

дифференциальных уравнений. Из предыдущих алгоритмов, тоько метод Ньютона подходит для 

решения систем нелинейных уравнений. 

Предположим, что мы имеем $n$ нелинейных уравнений, написанных в следующем виде:

$$
F_0(x_0,x_1,....,x_n)=0 \\
F_1(x_0,x_1,....,x_n)=0 \\
.... \\
.... \\
F_n(x_0,x_1,....,x_n)=0
$$

Эту систему можно записать в векторном виде:

$$
\vec{F}(\vec{x})=\vec{0}
$$

Мы следуем идеям метода Ньютона для одного уравнения с одной переменной:

аппроксимировать нелинейную $f(x)$ линейной функцией и найти корень этой функции. Когда

существует $n$ переменных, нам нужно приблизить вектор-функцию $F(x)$ некоторой линейной 

функцией $\vec{F}(\vec{x})=J\vec{x}+\vec{c}$, где $J$ матрица размера $nхn$ и $c$ - некоторый 

вектор длины $n$

Линейная аппроксимация методом Ньютона системы нелинейных алгебраических уравнений

записывается так:

$$
\vec{F}(\vec{x_{i+1}})\approx\vec{F}(\vec{x_i})+J(\vec{x_i})(\vec{x_{i+1}}-\vec{x_i})=\vec{0}
$$

до делать!!!!

#### Полный перебор (или метод грубой силы)



```python
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt

def brute_force_root_finder(f, a, b, n):    
    x = np.linspace(a, b, n)
    y = f(x)
    
    roots = []

    for i in range(n-1):
        if y[i] * y[i+1] < 0:
            root = x[i] - (x[i+1] - x[i]) / (y[i+1] - y[i]) * y[i]
            roots.append(root)

    return np.array(roots)

def f(x):
    y = np.exp(-x**2) * np.cos(4*x)
    return y
    
a = 0
b = 2
n = 1001

roots = brute_force_root_finder(f, a, b, n)
print(roots)

x = np.linspace(a, b, n)
y = f(x)

y0 = f(roots)

plt.plot(x, y, '-k', roots, y0, 'ob')

plt.grid(True)
plt.show()
```

```python
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt

def brute_force_optimizer(f, a, b, n):
    x = np.linspace(a, b, n)
    y = f(x)
    
    # поиск локальных экстремумов
    minima = []
    maxima = []

    for i in range(n-1):
        if y[i-1] < y[i] > y[i+1]:
            maxima.append(i)
        
        if y[i-1] > y[i] < y[i+1]:
            minima.append(i)
        
    # найдем глобальный максимум и минимум на интервале (a,b)    
    y_max_inner = max([y[i] for i in maxima])
    y_min_inner = min([y[i] for i in minima])

    # сравним со значениями функции на границах отрезка [a,b]
    
    if y[0] > y_max_inner:
        maxima.append(0)

    if y[len(x)-1] > y_max_inner:
        maxima.append(len(x)-1)

    if y[0] < y_min_inner:
        minima.append(0)

    if y[len(x)-1] < y_min_inner:
        minima.append(len(x)-1)

    x_min = [x[i] for i in minima]
    x_max = [x[i] for i in maxima]
    
    return np.array(x_min), np.array(x_max)


# зададим функцию, для которой будем искать экстремумы
def f(x):
    y = np.exp(-x**2) * np.cos(4*x)
    return y
    
a = 0
b = 2
n = 1001

x_min, x_max = brute_force_optimizer(f, a, b, n)
print('минимумы: ', x_min)
print('максимумы: ', x_max)

x = np.linspace(a, b, n)
y = f(x)

y_min = f(x_min)
y_max = f(x_max)

plt.plot(x, y, '-k', x_min, y_min, 'ob', x_max, y_max, 'or')

plt.grid(True)
plt.show()
```

#### Метод Ньютона нахождения корней нелинейных уравнений

```python
import sys
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt

def newton(f, df, x, eps, max_iterations=100):
    f_value = f(x)
    iteration_counter = 0

    while abs(f_value) > eps and iteration_counter < max_iterations:
        try:
            x = x - float(f_value) / df(x)
        except ZeroDivisionError:
            print('Error! - derivative zero for x = ', x) # не обращалась в ноль
            sys.exit(1)

        f_value = f(x)
        iteration_counter += 1 # счетчик итерации, что бы не было больше 100

        
    return x
    
# зададим функцию, для которой будем искать экстремумы
def f(x):
    y = np.exp(-x**2) * np.cos(4*x) # np программа для функции
    return y

def df(x):
    y = -2*x*np.exp(-x**2) * np.cos(4*x) - 4*np.exp(-x**2) * np.sin(4*x)
    return y
    
root = newton(f, df, 1.0, 1.0e-6, 100)
print(root)

a = 0
b = 2
n = 1001

x = np.linspace(a, b, n)
y = f(x)

plt.plot(x, y, '-k', root, f(root), 'ob')

plt.grid(True)
plt.show()
```

#### Метод секущих нахождения корней нелинейных уравнений

```python
import sys
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt

def secant(f, x0, x1, eps, max_iterations=100):
    f_x0 = f(x0)
    f_x1 = f(x1)
    
    iteration_counter = 0

    while abs(f_x1) > eps and iteration_counter < max_iterations:
        try:
            denominator = float(f_x1 - f_x0) / (x1 - x0)
            x = x1 - float(f_x1) / denominator
        except ZeroDivisionError:
            print('Error! - denominator zero for x = ', x)
            sys.exit(1)
        
        x0 = x1
        x1 = x

        f_x0 = f_x1
        f_x1 = f(x1)

        iteration_counter += 1

    return x

def f(x):
    return x**2 - 9

x0 = 1000
x1 = x0 - 1

root= secant(f, x0, x1, eps=1.0e-6)
print(root)

x = np.linspace(-10, 10, 1000)
y = f(x)

plt.plot(x, y, '-k', root, f(root), 'ob')

plt.grid(True)
plt.show()
```

#### Метод бисекции

```python
import sys
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt

def bisection(f, x_L, x_R, eps):
    f_L = f(x_L)

    if f_L * f(x_R) > 0:
        print('Функция на концах интервала одного знака')
        sys.exit(1)

    x_M = float(x_L + x_R)/2.0
    f_M = f(x_M)

    while abs(f_M) > eps:
        if f_L * f_M > 0:
            x_L = x_M
            f_L = f_M
        else:
            x_R = x_M
        
        x_M = float(x_L + x_R) / 2
        f_M = f(x_M)
    
    return x_M

def f(x):
    return x**2 - 9

root= bisection(f, 0, 1000, eps=1.0e-6)
print(root)

x = np.linspace(-10, 10, 1000)
y = f(x)

plt.plot(x, y, '-k', root, f(root), 'ob')

plt.grid(True)
plt.show()
```

### Решение системы нелинейных уравнений методом Ньютона

```python
import numpy as np
from numpy import cos, sin, pi, exp

def newton_system(F, J, x, eps, max_iterations=100):
    F_value = F(x)
    F_norm = np.linalg.norm(F_value, ord=2) # l2 norm of vector
    
    iteration_counter = 1
    
    while abs(F_norm) > eps and iteration_counter < max_iterations:
        delta = np.linalg.solve(J(x), -F_value)
        x = x + delta
        F_value = F(x)
        F_norm = np.linalg.norm(F_value, ord=2)
        iteration_counter += 1
    
    return x

def F(x):
    return np.array([x[0]**2 - x[1] + x[0]*cos(pi*x[0]),
                     x[0]*x[1] + exp(-x[1]) - x[0]**(-1)])

def J(x):
    return np.array([[2*x[0] + cos(pi*x[0]) - pi*x[0]*sin(pi*x[0]), -1],
                     [x[1] + x[0]**(-2), x[0] - exp(-x[1])]])

# точное решение (1, 0)
x = newton_system(F, J, x=np.array([2, -1]), eps=0.0001)
print(x)
```
